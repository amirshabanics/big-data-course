## بخش الف 

### داده را می‌خوانیم

```python
df = pd.read_csv("IMDB-Movie-Data.csv")
df
```

## بخش ب

### ژانر‌ها و بازیگران را از حالت رشته به آرایه‌ای از رشته‌ها تبدیل می‌کنیم

```python
df = df.assign(Genre=df.Genre.str.split(','), Actors=df.Actors.str.split(','))
df
```

### برای هر سطر به ازای هر ژانر یک سطر درست می‌کنیم

```python
genres = df.explode('Genre')
genres
```

### ستون جدید برای شمارش تعداد فیلم‌ها اضافه می‌کنیم
### جدول جدیدی براساس ژانر تولید می‌کنیم و تعداد فیلم‌ها و میانگین درآمد را محاسبه می‌کنیم

```python
genres['Count'] = 1
generes = genres.pivot_table(values=['Revenue (Millions)', 'Count'], index='Genre', aggfunc={'Revenue (Millions)': np.mean, 'Count': np.sum})
generes
```

## بخش ج

### هر دو سطر را باهمدیگر ادغام می‌کنیم. سپس سطر‌هایی که با خودشان ادغام شدند را کنار می‌گذاریم

```python
cross_df = df.merge(df, how='cross')
cross_df = cross_df[cross_df.Title_x != cross_df.Title_y]
```

### ستون جدیدی به نام بازیگران مشترک به آن اضافه می‌کنیم و تمام سطر‌هایی که بازیگر مشترک ندارند را دور می‌اندازیم

```python
cross_df['Intersect Actors'] = cross_df.apply(lambda row: ",".join(np.intersect1d(row.Actors_x, row.Actors_y).tolist()), axis=1)
cross_df = cross_df[cross_df['Intersect Actors'].str.len() > 0]
cross_df['Intersect Actors']
```

### ستون جدیدی به نام فیلم‌های ادغام‌شده درست می‌کنیم و لیست دو فیلمی که باهم ادغام شدند را می‌گذاریم

```python

cross_df['Joint Films'] = cross_df.apply(lambda row: f"{row.Title_x},{row.Title_y}", axis=1)
cross_df['Joint Films']
```

### به دلیل قابل هش نبودن لیست، ستون‌های جدید رو به صورت رشته ذخیره کردیم 

### حال براساس لیست بازیگر‌های مشترک، سطر‌ها را باهم ادغام می‌کنیم و لیست فیلم‌های ادغام شده را ادغام می‌کنیم

### برای اینکار یک تابع نوشتیم که سری‌هایی که بازیگر مشترک یکسان دارند را می‌گیرد و آن‌ها را با ویرگول به هم متصل می‌کند اما ممکن است در آن‌ها تکراری باشد. برای از بین بردن تکراری‌ها از تابع نامپای استفاده می‌کنیم و سپس دوباره آن‌ها را به رشته تبدیل می‌کنیم

```python
from functools import reduce


def join_films(series):
    films = reduce(lambda x, y: f"{x},{y}", series)
    return ",".join(np.unique(films.split(',')))


intersect_actors_in_films = cross_df.pivot_table(values=['Joint Films'], index=[
    'Intersect Actors'], aggfunc={'Joint Films': join_films})


intersect_actors_in_films
```
